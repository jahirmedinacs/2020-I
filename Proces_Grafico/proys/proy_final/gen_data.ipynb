```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
features = np.load("sevensegdataset.npy")
X = features[:, :-1]
Y = features[:, -1]
```

```python
for _ in range(5):
    id_x = int(np.floor(np.random.rand() * len(Y)))

    print(str(id_x))
    plt.imshow(X[id_x].reshape(28,28))
    plt.title(str(Y[id_x]))
    plt.show()
```
