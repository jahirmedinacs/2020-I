### Tiempo de Ejecucion

Consideracion Importante: Los metodos Locales usan un metodo de aplicacion de mascaras optimizado, que es aproximadamente 100 - 150 veces mas rapido que una aplicacion sin optimizar **(vectorizacion vs iteracion)**. Por esto se mostrara los promedios ajustando los resultados en un factor de 150.

```python
import pandas as pd
tiempos = pd.read_csv("./reformated_times.csv")
```

```python
tiempos
```

**Tiempos Promedio (sin ajuste)**

```python
tiempos.mean()
```

**Tiempos Promedio (con ajuste)**

```python
from copy import copy as copy
tiempos_ajuste = copy(tiempos)
tiempos_ajuste[tiempos_ajuste.columns[3:]] *= 150
tiempos_ajuste.mean()
```

### Discucion de Resultados

Es importante notar en las images resultado, que los metodos gloables , son exelentes para segmentar informacion de manera general, resaltando detalles (como en coins , que se resalta el volumen de los bordes) o marcando niveles en la escena (como en cameraman, donde los metodos gloables separan al individuo del fondo).

Sin embargo, metodos locales son mas eficientes con escenas donde existen multiples objetos pero bien diferenciados, con iluminacion uniforme y buena densidad de pixeles por imagen (como en pout, usando niblack).

Si bien cada metodo tiene su eficiencia computacional, debemos ver los resultados y no solo el tiempo, pues muchos metodos, como el isodata, son redudantes pero en algunos casos dan resultados excelentes (considerece la imagen2)
